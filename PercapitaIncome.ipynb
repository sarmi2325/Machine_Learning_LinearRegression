```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
d=pd.read_csv("canada_per_capita_income.csv")
d.head(10)
```

```python
d.isnull().sum()
```

```python
d.rename(columns={'per capita income (US$)':'pci'},inplace=True)

plt.scatter(d.year, d.pci,color='blue',marker='*')
```

```python
d
pci2 = d.pci.values.reshape(-1, 1)
d
```

```python
lm=LinearRegression()
lm.fit(d[['year']],pci2)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict([[2020]])
```

```python
plt.scatter(d.year, d.pci,color='blue',marker='*')
plt.xlabel("pci")
plt.ylabel("Year")
plt.plot(d.year,lm.predict(d[['year']]),color='red')
```
